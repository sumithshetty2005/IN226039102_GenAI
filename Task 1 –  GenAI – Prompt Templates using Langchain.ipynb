# Task 4: ChatPrompt Template System

```python
def get_chat_response(role_type, topic_name):
    roles = {
        "teacher": "You are a helpful teacher.",
        "interviewer": "You are a professional interviewer.",
        "motivator": "You are an inspiring coach."
    }

    chat_prompt = ChatPromptTemplate.from_messages([
        ("system", roles.get(role_type, "You are a helpful assistant.")),
        ("user", "{topic}")
    ])

    return chat_prompt.format_messages(topic=topic_name)

print("Chat Prompt:", get_chat_response(role_type='teacher', topic_name='Neural Networks'))
```

# Task 5: Input Validation Layer

```python
def validate_inputs(audience, tone):
    valid_audiences = ["beginner", "intermediate", "expert"]
    valid_tones = ["formal", "casual", "fun"]

    final_audience = audience if audience in valid_audiences else "beginner"
    final_tone = tone if tone in valid_tones else "formal"

    return final_audience, final_tone
```

# Task 6: Prompt Generator App

```python
def generate_prompt(topic, audience, tone, style):
    v_audience, v_tone = validate_inputs(audience, tone)

    styles = {
        "teaching": "Explain {topic} for {audience} in a {tone} teaching style.",
        "interview": "Prepare interview questions on {topic} for an {audience} level in a {tone} tone.",
        "storytelling": "Explain {topic} for {audience} in a {tone} storytelling style."
    }

    selected_tpl = styles.get(style, styles["teaching"])
    prompt_obj = PromptTemplate.from_template(selected_tpl)

    return prompt_obj.format(topic=topic, audience=v_audience, tone=v_tone)

print("App Output:", generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
```
